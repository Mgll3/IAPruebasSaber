```python
# Lectura archivos
df = pd.read_csv('train.csv')

# Revisión de primeros datos
df.head(5)
```

```python
# Ultimos datos
df.tail()
```

```python
total = df.shape[0]
alto = df[df['RENDIMIENTO_GLOBAL'] == 'alto'].shape[0]
alto_porcentaje = (alto/total)*100
medio_alto = df[df['RENDIMIENTO_GLOBAL'] == 'medio-alto'].shape[0]
medio_alto_porcentaje = (medio_alto/total)*100
medio = df[df['RENDIMIENTO_GLOBAL'] == 'medio-bajo'].shape[0]
medio_porcentaje = (medio/total)*100
bajo = df[df['RENDIMIENTO_GLOBAL'] == 'bajo'].shape[0]
bajo_porcentaje = (bajo/total)*100

print(f"Cantidad de Estudiantes con rendimiento Alto: {alto}, con un {alto_porcentaje:.2f}%" )

print(f"Cantidad de Estudiantes con rendimiento Medio-Alto: {medio_alto}, con un {medio_alto_porcentaje:.2f}%" )

print(f"Cantidad de Estudiantes con rendimiento Medio-bajo: {medio}, con un {medio_porcentaje:.2f}%" )

print(f"Cantidad de Estudiantes con rendimiento Bajo: {bajo}, con un {bajo_porcentaje:.2f}%" )
```

```python
# Frecuencia de estudiantes en cada programa academico
print(df['ESTU_PRGM_ACADEMICO'].value_counts())
```

```python
# Verificación de cantidad de valores nulos en cada columna
print(df.isnull().sum())
```

```python
# Cantidad de estudiantes de cada estrato
df['FAMI_ESTRATOVIVIENDA'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribución de Estrato de la Vivienda')
plt.xlabel('Estrato')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df['ESTU_HORASSEMANATRABAJA'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de Horas Semanales Trabajadas')
plt.xlabel('Horas Semanales Trabajadas')
plt.ylabel('Número de Estudiantes')
plt.show()
```
